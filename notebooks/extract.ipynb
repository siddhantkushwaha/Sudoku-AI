```python
import  cv2 as cv
import matplotlib.pyplot as plt
```

```python
path = '../data/train/image1078.jpg'
image = cv.cvtColor(cv.imread(path), cv.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
from clean import clean

image, _ = clean(image)
plt.imshow(image)
```

```python
from split import get_cells

cells = get_cells(image)
```

```python
def plot_cells(cells):
    fig, axs = plt.subplots(9, 9)
    
    for r, row in enumerate(cells, 0):
        for c, cell in enumerate(row, 0):
            axs[r, c].imshow(cell, cmap='gray')
            axs[r, c].axis('off')
            
plot_cells(cells)
```
